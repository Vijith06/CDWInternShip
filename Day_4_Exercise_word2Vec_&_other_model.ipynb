Word2Vec

```python
import gensim.downloader as api

# Download the pre-trained Word2Vec model
word2vec = api.load('word2vec-google-news-300')
print("Model loaded successfully!")
```

```python
from gensim.models import KeyedVectors
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import gensim.downloader as api


def preprocess(sentence):
    # Tokenize using basic splitting and convert to lowercase
    tokens = sentence.lower().split()
    # Keep only tokens present in Word2Vec vocabulary
    return [word for word in tokens if word in word2vec]

def sentence_to_vector(sentence):
    words = preprocess(sentence)
    if not words:
        return np.zeros(300)  # Return a zero vector if no valid words are found
    return np.mean([word2vec[word] for word in words], axis=0)

def calculate_similarity(sentence1, sentence2):
    vector1 = sentence_to_vector(sentence1)
    vector2 = sentence_to_vector(sentence2)
    similarity = cosine_similarity([vector1], [vector2])[0][0]
    return similarity

# Input sentences
sentence1 = "I love programming."
sentence2 = "Coding is my passion."

# Calculate and print similarity
similarity = calculate_similarity(sentence1, sentence2)
print(f"Cosine Similarity: {similarity:.4f}")
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load a pre-trained sentence embedding model (e.g., all-MiniLM-L6-v2)
model = SentenceTransformer('google-bert/bert-base-uncased')

# Input sentences
sentence1 = "I love programming."
sentence2 = "Coding is my passion."

# Encode sentences into embeddings
embedding1 = model.encode(sentence1)
embedding2 = model.encode(sentence2)

# Calculate cosine similarity
similarity = cosine_similarity([embedding1], [embedding2])[0][0]

print(f"Cosine Similarity: {similarity:.4f}")
```
